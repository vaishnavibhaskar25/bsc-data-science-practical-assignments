Question Set A

Question 1: Write a program to find the union, intersection, and difference of two sets entered by the user.

```python
set1 = set(input("Enter elements of first set (space separated): ").split())
set2 = set(input("Enter elements of second set (space separated): ").split())

print("Union:", set1 | set2)
print("Intersection:", set1 & set2)
print("Difference (Set1 - Set2):", set1 - set2)
```

Question 2: Write a program to check if a set is a subset of another set.

```python
set1 = set(input("Enter elements of first set: ").split())
set2 = set(input("Enter elements of second set: ").split())

if set1.issubset(set2):
    print("First set is a subset of second set")
else:
    print("First set is not a subset of second set")
```

Question 3: Write a program to remove duplicates from a list using sets

```python

lst = input("Enter list elements (space separated): ").split()

unique_list = list(set(lst))

print("List after removing duplicates:", unique_list)
```

Question 4: Write a program to find elements that are present in the first set but not in the second set.

```python
set1 = set(input("Enter elements of first set: ").split())
set2 = set(input("Enter elements of second set: ").split())

result = set1 - set2
print("Elements in first set but not in second:", result)
```

Question 3: Write a program to find the length of a set and check if a specific element exists in it.

```python
my_set = set(input("Enter elements of the set: ").split())

print("Length of set:", len(my_set))

element = input("Enter element to search: ")

if element in my_set:
    print("Element exists in the set")
else:
    print("Element does not exist in the set")
```

Question 1: Write a program that takes two sentences as input and finds:

```python
sent1 = input("Enter first sentence: ")
sent2 = input("Enter second sentence: ")

set1 = set(sent1.split())
set2 = set(sent2.split())

print("Common words:", set1 & set2)
print("Unique to first sentence:", set1 - set2)
print("Unique to second sentence:", set2 - set1)
```

Question 2: Write a program to perform all set operations (union, intersection, difference, symmetric difference) on three sets.


```python
set1 = set(input("Enter elements of first set: ").split())
set2 = set(input("Enter elements of second set: ").split())
set3 = set(input("Enter elements of third set: ").split())

print("Union:", set1 | set2 | set3)
print("Intersection:", set1 & set2 & set3)
print("Difference (Set1 - Set2 - Set3):", set1 - set2 - set3)
print("Symmetric Difference:", set1 ^ set2 ^ set3)
```
