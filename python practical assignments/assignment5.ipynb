1) Write a program to create a dictionary with roll number as key and name as value for 5 students. Display all key-value pairs.

```python
# Create empty dictionary
students = {}

# Take input for 5 students
for i in range(5):
    roll = int(input("Enter roll number: "))
    name = input("Enter student name: ")
    students[roll] = name

# Display dictionary
print("Student Dictionary:")
for roll, name in students.items():
    print(roll, "=>", name)
```

2) Write a program to count the frequency of each word in a given sentence using a dictionary.

```python
sentence = input("Enter a sentence: ")
words = sentence.split()

freq = {}
for word in words:
    if word in freq:
        freq[word] += 1
    else:
        freq[word] = 1

print("Word Frequency:", freq)
```

3) Write a program to merge two dictionaries entered by the user.


```python
dict1 = {}
dict2 = {}

n1 = int(input("Enter number of elements for first dictionary: "))
for i in range(n1):
    key = input("Enter key: ")
    value = input("Enter value: ")
    dict1[key] = value

n2 = int(input("Enter number of elements for second dictionary: "))
for i in range(n2):
    key = input("Enter key: ")
    value = input("Enter value: ")
    dict2[key] = value

merged_dict = {**dict1, **dict2}
print("Merged Dictionary:", merged_dict)
```

4) Write a program to find the key with the maximum value in a dictionary.


```python
data = {}
n = int(input("Enter number of elements: "))

for i in range(n):
    key = input("Enter key: ")
    value = int(input("Enter value: "))
    data[key] = value

max_key = max(data, key=data.get)
print("Key with maximum value:", max_key)
```

<h1>Question Set B

1) Write a program to remove a key from a dictionary and handle the case when key doesn't exist.

```python
data = {'a': 10, 'b': 20, 'c': 30}
key = input("Enter key to remove: ")

if key in data:
    del data[key]
    print("Updated Dictionary:", data)
else:
    print("Key not found")
```

2) Write a program to sort a dictionary by its values in descending order.

```python
data = {'a': 10, 'b': 50, 'c': 30}

sorted_dict = dict(sorted(data.items(), key=lambda x: x[1], reverse=True))
print("Sorted Dictionary:", sorted_dict)
```

```python
keys = ['a', 'b', 'c']
values = [10, 20, 30]

dictionary = dict(zip(keys, values))
print("Created Dictionary:", dictionary)
```

1) Write a program to find all keys with the same value in a dictionary.


```python
data = {'a': 10, 'b': 20, 'c': 10, 'd': 20}

value = int(input("Enter value to find keys: "))
keys = []

for k, v in data.items():
    if v == value:
        keys.append(k)

print("Keys with same value:", keys)
```

2) Write a program to create a dictionary where keys are numbers from 1 to n and values are their cubes.

```python
n = int(input("Enter value of n: "))

cube_dict = {}
for i in range(1, n + 1):
    cube_dict[i] = i ** 3

print("Dictionary of cubes:", cube_dict)
```
