# 1) write a program to calculate the average of numbers in a given list

```python

numbers = [10, 15, 20, 5]

total = 0
for num in numbers:
    total = total + num

average = total / len(numbers)

print("Average =", average)
```

<h3> 2)write a program which accepts 6 integer values and print duplicates if any of the value entered are duplicates otherwise print  all unique

```python
numbers = []

print("Enter 6 integers:")
for i in range(6):
    n = int(input())
    numbers.append(n)

duplicates = []

for num in numbers:
    if numbers.count(num) > 1 and num not in duplicates:
        duplicates.append(num)

if len(duplicates) > 0:
    print("Duplicate values are:", duplicates)
else:
    print("All values are unique:", numbers)
```

<h3>4) write a program which finds sum of dights of anumber

```python
n = int(input(" enter a number: "))
sum_digits = 0
while n > 0:
    digit = n % 10
    sum_digits = sum_digits + digit
    n = n // 10

print("sum of digits =", sum_digits)
```

```python
n = int(input("Enter number of terms: "))

a = 0
b = 1

print("Fibonacci Series:")
print(a, b, end=" ")

for i in range(2, n):
    c = a + b
    print(c, end=" ")
    a = b
    b = c
```

<h3>1) Write a program which accepts and integer value n display all prime numbers till n

```python
n = int(input("Enter the value of n: "))

print("Prime numbers till", n, "are:")

for num in range(2, n + 1):
    for i in range(2, num):
        if num % i == 0:
            break
    else:
        print(num, end=" ")
```

<h3> 1) write a sequential search funcation which searches an item in a sorted list. The function should return the index of element to be searched in the list

```python
def sequential_search(lst, key):
    for i in range(len(lst)):
        if lst[i] == key:
            return i
        elif lst[i] > key:   # list is sorted
            return -1
    return -1


# Main Program
numbers = [10, 20, 30, 40, 50, 60]

key = int(input("Enter element to search: "))

index = sequential_search(numbers, key)

if index != -1:
    print("Element found at index", index)
else:
    print("Element not found")
```
