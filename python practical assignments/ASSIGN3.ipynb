Question 1: Write a program to find the maximum and minimum element in a tuple without using max() and min() functions.

```python
tup = (3, 7, 1, 9, 4)

maximum = tup[0]
minimum = tup[0]

for num in tup:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

print("Maximum element:", maximum)
print("Minimum element:", minimum)
```

Question 2: Write a program to count the occurrence of each element in a tuple.

```python
tup = (1, 2, 3, 2, 1, 3, 2)

count = {}

for item in tup:
    if item in count:
        count[item] += 1
    else:
        count[item] = 1

for key in count:
    print(key, ":", count[key])
```

Question 3: Write a program to convert a tuple of strings into a single string.

```python
tup = (1, 2, 3, 4, 5, 6)

total = 0

for num in tup:
    if num % 2 == 0:
        total += num

print("Sum of even numbers:", total)
```

Question 5: Write a program to remove duplicates from a tuple and return a new tuple with unique elements.

```python
tup = (1, 2, 3, 2, 1, 4)

unique_list = []

for item in tup:
    if item not in unique_list:
        unique_list.append(item)

unique_tuple = tuple(unique_list)
print("Tuple without duplicates:", unique_tuple)
```

Question 1: Write a program to find the second largest element in a tuple.

```python
tup = (10, 20, 5, 40, 30)

largest = second_largest = -1

for num in tup:
    if num > largest:
        second_largest = largest
        largest = num
    elif num > second_largest and num != largest:
        second_largest = num

print("Second largest element:", second_largest)
```

Question 2: Write a program to sort a tuple of tuples based on the second element.

```python
tup = (1, 2, 3, 4)

is_unique = True

for item in tup:
    if tup.count(item) > 1:
        is_unique = False
        break

if is_unique:
    print("All elements are unique")
else:
    print("Elements are not unique")
```

Question 4: Write a program to swap the first and last element of a tuple.

```python
up = (10, 20, 30, 40)

new_tup = (tup[-1],) + tup[1:-1] + (tup[0],)

print("Tuple after swapping:", new_tup)
```

Question 1: Write a program to find common elements between two tuples.

```python
tup1 = (1, 2, 3, 4)
tup2 = (3, 4, 5, 6)

common = []

for item in tup1:
    if item in tup2 and item not in common:
        common.append(item)

print("Common elements:", tuple(common))
```

Question 2: Write a program to create a tuple of squares of numbers from 1 to n.

```python
n = int(input("Enter value of n: "))

squares = []

for i in range(1, n + 1):
    squares.append(i * i)

result = tuple(squares)
print("Tuple of squares:", result)
```
