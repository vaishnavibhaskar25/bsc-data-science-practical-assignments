Question 1: Write a program to count the number of vowels in a given string.


```python
# Take input from the user
text = input("Enter a string: ")

# Define vowels
vowels = "aeiouAEIOU"
count = 0

# Check each character
for ch in text:
    if ch in vowels:
        count += 1

# Print result
print("Number of vowels:", count)
```

Question 2: Write a program to check if a string is a palindrome (reads the same forwards and backwards).

```python
# Take input from the user
text = input("Enter a string: ")

# Reverse the string using loop
reverse_text = ""
for ch in text:
    reverse_text = ch + reverse_text

# Check palindrome
if text == reverse_text:
    print("It is a palindrome")
else:
    print("It is not a palindrome")
```

Question 4: Write a program to count the frequency of each character in a string.

```python
# Take input from the user
text = input("Enter a string: ")

# Empty dictionary
frequency = {}

# Count characters
for ch in text:
    if ch in frequency:
        frequency[ch] += 1
    else:
        frequency[ch] = 1

# Print result
for key in frequency:
    print(key, ":", frequency[key])
```

Question 5: Write a program to remove all whitespace from a string.

```python
# Take input from the user
text = input("Enter a string: ")

result = ""

# Remove spaces
for ch in text:
    if ch != " ":
        result += ch

print("String without spaces:", result)
```

Question 1: Write a program to find the first non-repeating character in a string.

```python
# Take input from user
text = input("Enter a string: ")

# Dictionary to store frequency
count = {}

# Count frequency of each character
for ch in text:
    if ch in count:
        count[ch] += 1
    else:
        count[ch] = 1

# Find first non-repeating character
for ch in text:
    if count[ch] == 1:
        print("First non-repeating character is:", ch)
        break
```

Question 2: Write a program to check if two strings are anagrams.

```python
str1 = input("Enter first string: ")
str2 = input("Enter second string: ")

# Remove spaces and convert to lowercase
str1 = str1.replace(" ", "").lower()
str2 = str2.replace(" ", "").lower()

if sorted(str1) == sorted(str2):
    print("Strings are anagrams")
else:
    print("Strings are not anagrams")
```

Question 3: Write a program to convert a string to title case without using the title() method.

```python
text = input("Enter a string: ")

words = text.split()
result = ""

for word in words:
    result += word[0].upper() + word[1:] + " "

print("Title Case String:", result.strip())
```

Question 1: Write a program to find the longest word in a sentence.

```python
sentence = input("Enter a sentence: ")

words = sentence.split()
longest = words[0]

for word in words:
    if len(word) > len(longest):
        longest = word

print("Longest word:", longest)
```

Question 2: Write a program to check if a string contains only digits, only alphabets, or both.

```python
text = input("Enter a string: ")

if text.isdigit():
    print("String contains only digits")
elif text.isalpha():
    print("String contains only alphabets")
else:
    print("String contains both digits and alphabets")
```
