## Question Set A
### 1. Function to find maximum of three numbers

```
def max_of_three(a, b, c):
    return max(a, b, c)

print(max_of_three(10, 25, 15))
```

### 2. Function to check if a number is prime

```
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

print(is_prime(7))
```

### 3. Factorial using iterative and recursive methods

```
def factorial_iterative(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    return fact

def factorial_recursive(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial_recursive(n - 1)

print("Iterative:", factorial_iterative(5))
print("Recursive:", factorial_recursive(5))
```

### 4. Function to return unique elements from a list

```
def unique_list(lst):
    return list(set(lst))

print(unique_list([1, 2, 2, 3, 4, 4]))
```

### 5. Function to calculate sum of all even numbers in a list

```
def sum_of_evens(lst):
    return sum(i for i in lst if i % 2 == 0)

print(sum_of_evens([1, 2, 3, 4, 5, 6]))
```

## Question Set B
### 1. Temperature conversion

```
def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

def fahrenheit_to_celsius(f):
    return (f - 32) * 5/9

print(celsius_to_fahrenheit(25))
print(fahrenheit_to_celsius(77))
```

### 2. Count vowels and consonants in a string

```
def count_vowels_consonants(s):
    vowels = 'aeiouAEIOU'
    v = c = 0
    for ch in s:
        if ch.isalpha():
            if ch in vowels:
                v += 1
            else:
                c += 1
    return {'vowels': v, 'consonants': c}

print(count_vowels_consonants("Hello World"))
```

### 3. Function with variable number of arguments to find average

```
def average(*args):
    return sum(args) / len(args)

print(average(10, 20, 30, 40))
```

### 4. Lambda function to find maximum of three numbers

```
max_three = lambda a, b, c: max(a, b, c)
print(max_three(5, 15, 10))
```

## Question Set C
### 1. Recursive function to find sum of digits

```
def sum_of_digits(n):
    if n == 0:
        return 0
    return n % 10 + sum_of_digits(n // 10)

print(sum_of_digits(1234))
```

### 2. Count even and odd numbers in a list

```
def count_even_odd(lst):
    result = {'even': 0, 'odd': 0}
    for i in lst:
        if i % 2 == 0:
            result['even'] += 1
        else:
            result['odd'] += 1
    return result

print(count_even_odd([1, 2, 3, 4, 5]))
```
